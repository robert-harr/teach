# Capacitor Edge Effects

Plot the field and equipotentials at the edge of a circular, parallel-plate capacitor. These results are obtained using conformal mapping.
There is another result obtained by Landau and Lifshitz which differs slightly from this one.


```python

import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,15))

# Draw equipotentials
V = np.linspace(-5., 5., 500)
for U in np.linspace(-np.pi*1.j, np.pi*1.j, 15):
    X = -np.exp(V-U) - (V+1) + U
    ax.plot(np.real(X), np.imag(X), color='c')

# Draw field lines
U = np.linspace(-np.pi*1.j, np.pi*1.j, 500)
for V in np.linspace(-4.5, 1.0, 9):
    X = -np.exp(V-U) - (V+1) + U
    ax.plot(np.real(X), np.imag(X), color='r')

# Draw the capacitor plates
ax.hlines(np.pi, 0., 4., linewidth=6, color='k')
ax.hlines(-np.pi, 0., 4., linewidth=6, color='k')

ax.set(xlim=(-5, 4), ylim=(-6, 6))
plt.show()
```
